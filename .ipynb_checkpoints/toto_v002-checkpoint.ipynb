```python
# 空のDataFrameを作成
url = 'http://www.football-lab.jp/ka-f/match/?year=2012'
data = pd.io.html.read_html(url)[3]
df_raw = DataFrame(data)
df = pd.DataFrame(columns = df_raw.iloc[0,2:])
df['year']=''
df['league']=''
df['url']=''

# 対戦結果データのスクレイピング
for obs_tm in tm_list:    
    for i_year in range(6):

    # 対戦情報を取得
        print(obs_tm)
        print(i_year+2012)
        year = i_year+2012
        url = 'http://www.football-lab.jp/{}/match/?year={}'.format(obs_tm, year)
        data = pd.io.html.read_html(url)[3]

        df_raw = DataFrame(data)
        df2=df_raw.iloc[1:,2:]
        df2.columns = df_raw.iloc[0,2:]
        df2['year']=year
        df2['url']=url

    #リーグ情報を取得
        url_league = 'http://www.football-lab.jp/ka-f/?year={}'.format(year)

        # データを取得
        result = requests.get(url)
        c = result.content
        soup = BeautifulSoup(c, "html.parser")
        league = soup.find("div", attrs={"class": "unit", "id": "teamName"}).em.string
        df2['league']=league

        df = df.append(df2)
df
# df
```
